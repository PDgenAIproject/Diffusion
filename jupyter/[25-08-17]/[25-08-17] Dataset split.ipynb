```python
import argparse, os, sys, datetime, glob, importlib, csv
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split
```

### split whole dataset

##### Lt_SLIT

```python


csv_path = "/camin1/chlee/Data/cornea_image/segmented/Lt/slit/label.csv"
img_dir = "/camin1/chlee/Data/cornea_image/segmented/Lt/slit/whole"
output_dir = "/camin1/chlee/Data/cornea_image/segmented/Lt/slit/split"   # output 폴더는 분리 추천

# CSV 읽기
df = pd.read_csv(csv_path)

# 1차: train(80%) + temp(20%) 로 나누기
train_df, val_df = train_test_split(
    df,
    test_size=0.4,               # 60% train → 나머지 40%가 temp
    stratify=df['label'],
    random_state=42
)


split_dict = {
    'train': train_df,
    'val': val_df
}

# 폴더로 이미지 복사
for split, sub_df in split_dict.items():
    for _, row in sub_df.iterrows():
        filename = row['filename']
        label = str(row['label'])
        
        target_folder = os.path.join(output_dir, split, label)
        os.makedirs(target_folder, exist_ok=True)
        
        src = os.path.join(img_dir, filename)
        dst = os.path.join(target_folder, filename)
        
        if os.path.exists(src):
            shutil.copy(src, dst)
        else:
            print(f"파일 없음: {src}")

print("완료!")
```
